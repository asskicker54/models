```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('./data/adult.csv')
df.head()
```

```python
df = df.replace('?', np.nan)
```

```python
df.dropna()
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
del df['education']
```

```python
df['sex'] = df['sex'].replace({'Female': 0, 'Male': 1})
df['race'] = df['race'].replace({'White': 0, 'Black': 1})
y = df['income'].replace({'<=50K': 0, '>50K': 1})
X = df.drop('income', axis=1)
```

```python
X_transform = pd.get_dummies(X)
X_transform
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_transform, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import classification_report 

model = LogisticRegression() 
model.fit(X_train, y_train)  
y_pred = model.predict(X_test) 

print(classification_report(y_test, y_pred))
```

#### Нормализация

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_normalized = scaler.fit_transform(X_transform)
print(X_normalized)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size = 0.3, random_state = 42)
```

```python
lr_norm = LogisticRegression()
lr_norm.fit(X_train, y_train)
y_pred = lr_norm.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()
X_std = scale.fit_transform(X_transform)

X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size = 0.3, random_state = 42)

lr_std = LogisticRegression()
lr_std.fit(X_train, y_train)
y_pred = lr_std.predict(X_test)

print(classification_report(y_test, y_pred))
```

При стандартизации качество обучения модели получилось самым лучшим.

```python
from sklearn.model_selection import GridSearchCV

parameters = {'C':[0.01, 0.1, 0.5, 1, 5, 10, 50, 100, 500, 1000]}

logit = LogisticRegression(max_iter = 1000)

grid = GridSearchCV(logit, parameters)

grid.fit(X_transform, y)

grid.best_estimator_
```

#### Выводы:

* При стандартизации данные масштабируются к среднему значению
* При нормализации данные приводятся к значениям 0 или 1

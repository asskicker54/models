Постройте изображение с настройками по умолчанию в Matplotlib:


```python
def sinplot(flip = 1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i*.5)*(7 - i)*flip)
sinplot()
plt.show()
```

Измените тот же график на значения по умолчанию в Seaborn, используя функцию set():


```python
def sinplot(flip = 1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i*.5)*(7 - i)*flip)
sb.set()
sinplot()
plt.show()
```

Применим стиль "whitegrid"

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i*.5)*(7 - i)*flip)
sb.set_style("whitegrid")
sinplot()
plt.show()
```

Укажите разницу между двумя рисунками (примеры 4 и 5).

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i*.5)*(7 - i)*flip)
sb.set_style("white")
sinplot()
sb.despine()
plt.show()
```

Изменение значений любого параметра изменит стиль графика

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sb.set_style("darkgrid", {'axes.axisbelow': False}) # Изменили здесь
sinplot()
sb.despine()
plt.show()
```
